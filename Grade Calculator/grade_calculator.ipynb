```python
#PROJECT 2 ----Student Grade Calculator----
```

```python
# Stuent name Function
def get_student_name():
    name = input("Enter student name: ").strip()
    return name if name else "Student"
```

```python
# Marks Fuction
def get_valid_marks():
    while True:
        raw = input("Enter marks (0-100): ").strip()
        try:
            marks = float(raw)
        except ValueError:
            print("Invalid input — please enter a number between 0 and 100.")
            continue
        if marks.is_integer():
            marks = int(marks)
        if 0 <= marks <= 100:
            return marks
        print("Marks must be between 0 and 100. Try again.")
```

```python
# Calculating Grades
def calculate_grade(marks):
    if 90 <= marks <= 100:
        return "A"
    elif 80 <= marks <= 89:
        return "B"
    elif 70 <= marks <= 79:
        return "C"
    elif 60 <= marks <= 69:
        return "D"
    else:
        return "F"
```

```python
# Encouraging Message
def encouraging_message(grade):
    if grade == "A":
        return "Excellent! Outstanding performance! "
    elif grade == "B":
        return "Very Good! Keep it up! "
    elif grade == "C":
        return "Good effort — a little more focus will help. "
    elif grade == "D":
        return "You're getting there — ask for help and practice more. "
    else:  # F
        return "Don't be discouraged — review topics, ask questions, and try again. You can improve! "
```

```python
# Printing Result
def print_result(name, marks, grade, message):
    print(f"\nRESULT FOR {name.upper()}:")
    print(f"Marks: {marks}/100")
    print(f"Grade: {grade}")
    print(f"Message: {message}")
```

```python
# Main Function
def main():
    # Step 1 & 2: Get inputs
    name = get_student_name()
    marks = get_valid_marks()

    # Step 3: Grading logic
    grade = calculate_grade(marks)

    # Step 4: Encouraging message
    message = encouraging_message(grade)

    # Final output
    print_result(name, marks, grade, message)

if __name__ == "__main__":
    main()
```
